```python
import os

DATASET_DIR = "D:\VISUAL CODE\Microsoft VS Code\Eksperimen_SML_Batara/plantvillage_raw"

classes = sorted(os.listdir(DATASET_DIR))

print("Jumlah kelas:", len(classes))
```

```python
import pandas as pd

data = []

for cls in classes:

    total = len(
        os.listdir(
            os.path.join(DATASET_DIR, cls)
        )
    )

    data.append([cls,total])

df = pd.DataFrame(
    data,
    columns=["Class","Count"]
)

df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,6))

df.sort_values(
    by="Count",
    ascending=False
).plot(
    x="Class",
    y="Count",
    kind="bar"
)

plt.show()
```

```python
from PIL import Image
import random

plt.figure(figsize=(10,10))

for i in range(9):

    cls = random.choice(classes)

    file = random.choice(
        os.listdir(
            os.path.join(DATASET_DIR, cls)
        )
    )

    img = Image.open(
        os.path.join(
            DATASET_DIR,
            cls,
            file
        )
    )

    plt.subplot(3,3,i+1)
    plt.imshow(img)
    plt.title(cls)
    plt.axis("off")

plt.show()
```

```python
from PIL import Image
import shutil
import os

INPUT_DIR = "D:\VISUAL CODE\Microsoft VS Code\Eksperimen_SML_Batara/plantvillage_raw"
OUTPUT_DIR = "plantvillage_preprocessing"

os.makedirs(
    OUTPUT_DIR,
    exist_ok=True
)

for cls in os.listdir(INPUT_DIR):

    os.makedirs(
        os.path.join(
            OUTPUT_DIR,
            cls
        ),
        exist_ok=True
    )

    for file in os.listdir(
        os.path.join(INPUT_DIR, cls)
    ):

        try:

            img = Image.open(
                os.path.join(
                    INPUT_DIR,
                    cls,
                    file
                )
            )

            img.verify()

            shutil.copy(
                os.path.join(
                    INPUT_DIR,
                    cls,
                    file
                ),
                os.path.join(
                    OUTPUT_DIR,
                    cls,
                    file
                )
            )

        except:
            pass
```
